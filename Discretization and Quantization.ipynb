```python
import numpy as np
import matplotlib.pyplot as plt

x_1 = np.linspace(0,2*np.pi,100)
x_2 = np.linspace(0,2*np.pi,100)
y_1 = np.sin(x_1)
plt.plot(x_1,y_1,color = 'r')

plt.grid()
plt.xlabel('t')
plt.ylabel('sin(t)')
plt.xlim(0,6.3)
plt.ylim(-1.05,1.05)
plt.axhline(y=0, color='b')

plt.show()
```

```python
#Let say we want 20 number of samples
```

```python
Ts = np.pi/10
samples = 20
x = np.zeros(20)
for n in range(samples):
    x[n] = np.sin(n*(Ts)) 
```

```python
print(x)
x_min = min(x)
x_max = max(x)
print(x_min)
print(x_max)
```

```python
x_3 = np.linspace(0,2*np.pi,20)
```

```python
#plt.axhline(y=0, color='b')
for m in range(20):
    plt.ylim(-2,2)
    plt.plot(m*Ts,x[m],'o')
    plt.plot((m*Ts,m*Ts),(0,x[m]))
plt.axhline(y=0, color='b')
plt.xlabel('Time')
plt.ylabel('Discretized Signal')
```

```python
#Now to do quantization in amplitude domain:
##Suppose we are dividing the amplitude region into 7 samples(3-bit computer)
```

```python
N = 32
import math as math
Xq = np.zeros(20)
for i in range(20):
    Xq[i] =math.floor(((x[i] - x_min)/(x_max - x_min))*((2**N)-1))
```

```python
print(Xq)
```

```python
for m in range(20):
    plt.ylim(-20,2**(N)+300)
    plt.plot(m*Ts,Xq[m],'o')
    plt.plot((m*Ts,m*Ts),(Xq[0],Xq[m]))
plt.axhline(y=Xq[0], color='b')
plt.xlabel('Time')
plt.ylabel('Quantized & Discretized Signal')
#plt.label("Quantized and Discretized signal obtained from sinusoidal wave in 32-bit computer")
```
